```python
import os
import time
import datetime
import pandas as pd
from pygooglenews import GoogleNews

# 결과 저장 디렉토리 생성
BASE_DIR = "results/news"
os.makedirs(BASE_DIR, exist_ok=True)

CSV_FILE_PATH = os.path.join(BASE_DIR, "AAPL_news.csv")

def fetch_daily_news(query, start_date, end_date, max_per_day=5):
    """Google News에서 날짜별 뉴스 검색"""
    gn = GoogleNews(lang='en', country='US')  # 영어 뉴스 검색
    all_articles = []
    current = start_date

    while current <= end_date:
        date_str = current.strftime("%Y-%m-%d")
        next_day = (current + datetime.timedelta(days=1)).strftime("%Y-%m-%d")

        print(f"📅 Fetching articles for {date_str}...")

        # 날짜 필터 적용 (when 대신 after ~ before 사용)
        search_query = f"{query} after:{date_str} before:{next_day}"
        search_result = gn.search(search_query)
        entries = search_result.get('entries', [])

        valid_articles = []
        news_id = 1  # 뉴스 번호 (1부터 시작)

        for a in entries:
            if len(valid_articles) >= max_per_day:
                break  # 5개를 초과하면 중단

            title = a.get('title', 'No Title')

            # 유효한 기사 저장
            valid_articles.append({
                "Date": date_str,
                "News_ID": news_id,
                "Title": title
            })
            news_id += 1  # 뉴스 번호 증가

            time.sleep(0.5)  # 과부하 방지를 위한 딜레이

        # 최종 기사 추가
        all_articles.extend(valid_articles[:max_per_day])

        current += datetime.timedelta(days=1)
        time.sleep(1)  # 하루 단위 요청 사이 딜레이

    return pd.DataFrame(all_articles)

def save_news_to_csv(df):
    """뉴스 데이터를 하나의 CSV 파일에 저장 (누적 저장 방식)"""
    if os.path.exists(CSV_FILE_PATH):
        df.to_csv(CSV_FILE_PATH, mode='a', header=False, index=False, encoding="utf-8-sig")
    else:
        df.to_csv(CSV_FILE_PATH, mode='w', header=True, index=False, encoding="utf-8-sig")
    print(f"✅ Saved {len(df)} articles to {CSV_FILE_PATH}")

if __name__ == "__main__":
    # 설정값
    ticker = "AAPL"  # Apple의 실제 Ticker는 AAPL입니다!
    query = "Apple"  # 검색어 (영어)
    start_date = datetime.date(2020, 1, 1)
    end_date = datetime.date(2023, 12, 31)

    print(f"▶️ Fetching news for '{query}' from {start_date} to {end_date}...")
    df_news = fetch_daily_news(query, start_date, end_date)

    # CSV 파일에 저장
    save_news_to_csv(df_news)

    print("✅ All news data saved successfully!")
```
